```python
import pandas as pd
df=pd.read_csv('IRIS_DATASET.csv')
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=40)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
dct=DecisionTreeClassifier(criterion='entropy',random_state=1,max_depth=None)
dct.fit(x_train,y_train)
y_pred=dct.predict(x_test)
print('Prediction of [7,3.2,4.7,1.4] : ',dct.predict([[7,3.2,4.7,1.4]]))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,recall_score,precision_score,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
cm=confusion_matrix(y_test,y_pred)
print('Confusion Matrix\n',cm)
cm_display=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['setosa','versicolor','virginica'])
cm_display.plot()
print('Accuracy Score = ',accuracy_score(y_test,y_pred)*100,'\nF1 Score = ',f1_score(y_test,y_pred,average='macro'),'\nRecall Score = ',recall_score(y_test,y_pred,average='macro'),'\nPrecision Score = ',precision_score(y_test,y_pred,average='macro'))
```
