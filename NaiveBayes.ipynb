```python
from sklearn.datasets import load_iris
df=load_iris()
x=df.data
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
from sklearn.naive_bayes import GaussianNB
naive=GaussianNB()
naive.fit(x_train,y_train)
y_pred=naive.predict(x_test)
print('Prediction of [4.4,2.9,1.4,0.2] : ',df.target_names[naive.predict([[4.4,2.9,1.4,0.2]])])
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,recall_score,precision_score,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
cm=confusion_matrix(y_test,y_pred)
print('Confusion Matrix\n',cm)
cm_display=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['setosa','versicolor','virginica'])
cm_display.plot()
print('Accuracy Score = ',accuracy_score(y_test,y_pred)*100,'\nF1 Score = ',f1_score(y_test,y_pred,average='macro'),'\nRecall Score = ',recall_score(y_test,y_pred,average='macro'),'\nPrecision Score = ',precision_score(y_test,y_pred,average='macro'))
```
