```python
from sklearn.datasets import load_iris
df=load_iris()
x=df.data[:,:2]
y=df.target
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=4,n_init='auto',random_state=0)
kmeans.fit(x)
y_pred=kmeans.predict(x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1],label='DataPoints',cmap='viridis',c=y_pred,edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,label='Centroids')
plt.legend()
plt.show()
```
